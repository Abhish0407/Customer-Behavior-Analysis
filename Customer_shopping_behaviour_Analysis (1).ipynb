**Problem Statement**

A leading retail company wants to better understand its customers' shopping behavior in order to improve sales, customer satisfaction, and long-term loyalty. The management team has noticed changes in purchasing patterns across demographics, product categories, and sales channels (online vs. offline). They are particularly interested in uncovering which factors, such as discounts, reviews, seasons, or payment preferences, drive consumer decisions and repeat purchases.

You are tasked with analyzing the company's consumer behavior dataset to answer the following overarching business question:

*"How can the company leverage consumer shopping data to identify trends, improve customer engagement, and optimize marketing and product strategies?"*

```python
# We start off by loading the dataset and importing all the required libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv('F:\DA-PORTFOLIO\customer_shopping_behavior.csv')
```

```python
# To find the number of null values we have in our dataset
df.isnull().sum()
```

*Choosing median over mean to treat missing data because median is more robust and are less sensitive to extreme values.*

```python
# We will calculate & impute the median value of their respective categories
# (for ex.-  If a clothing item has  a missing value, we impute it with the median value of clothing category only, so that we don't get ambigious result)

df["Review Rating"] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df
```

```python
df.isnull().sum()

# now no more null values left.
```

```python
# Transforming the column name to snake casing as it is easier to code and easier to refernce in future cells

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create a New column age_group
labels = ['Young Adult', "Adult", "Middle-aged", "Senior"]
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# Creating a new column purchase_frequency_days (as it is harder to work with string type data in Frequency of purchase column)

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
    }
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
# We are assuming that if promo code is applied then automatically discounts are applied
# To confirm this in the dataset we must check the corresponding values

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Since the result came out to be true, We would be dropping the promo code column as it is of no use now.

df = df.drop('promo_code_used', axis = 1)
```
